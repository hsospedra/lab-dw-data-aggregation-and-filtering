```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df.columns = (df.columns.str.lower().str.replace(" ", "_"))
```

```python
df_low_claim_yes = df[(df['total_claim_amount'] < 1000) &(df["response"] == "Yes")]

df_low_claim_yes.head()
```

The filtered dataset includes customers with total claim amounts below 1,000 who responded positively to the marketing campaign. These customers typically show moderate to high customer lifetime values while maintaining low claim costs, suggesting they may represent lower-risk and more profitable profiles for the company.

```python
df_yes = df[df["response"] == "Yes"]
```

```python
avg_premium = (df_yes.groupby(["policy_type", "gender"])["monthly_premium_auto"].mean().round(2).reset_index())

avg_premium
```

Among customers who responded “Yes”, average monthly premiums vary by policy type. Personal Auto policies show the highest average premiums, followed by Corporate Auto and Special Auto. Differences between genders within each policy type are relatively small, indicating that policy type has a stronger influence on premium levels than gender.

```python
avg_clv = (df_yes.groupby(["policy_type", "gender"])["customer_lifetime_value"].mean().round(2).reset_index())

avg_clv
```

Customer lifetime value also varies more significantly by policy type than by gender. Personal Auto policies tend to show higher average CLV, particularly among female customers, while Corporate Auto and Special Auto display more moderate values. Gender differences exist but are not consistent across policy types.

```python
avg_claims = (df_yes.groupby(["policy_type", "gender"])["total_claim_amount"].mean().round(2).reset_index())

avg_claims
```

When comparing average customer lifetime value with average total claim amounts, no segment shows exceptionally high claim costs. Segments combining relatively higher CLV with similar or lower claim amounts may be considered more profitable or lower-risk for the company.

```python
customers_by_state = (df.groupby("state")["customer"].count().reset_index(name="num_customers"))

states_over_500 = customers_by_state[customers_by_state["num_customers"] > 500]

states_over_500
```

Customer counts are concentrated in a small number of states. California has the largest customer base, followed by Oregon and Arizona. This indicates a clear geographic clustering of customers, which may be relevant for targeted marketing or regional resource allocation.

```python
clv_stats = (df.groupby(["education", "gender"])["customer_lifetime_value"].agg(["min", "median", "max"]).round(2).reset_index())

clv_stats
```

Customer lifetime value varies more strongly by education level than by gender. Customers with “High School or Below” education show the highest median CLV values across both genders, while customers with doctoral education present the lowest median CLV. Gender differences within each education level are relatively small compared to the variation observed across education levels.

The median CLV provides a robust measure of central tendency, reducing the influence of extreme values observed in the maximum CLV across several education groups.

```python
#5.
df["effective_to_date"]
```

```python
df["effective_to_date"] = pd.to_datetime(df["effective_to_date"],format="%m/%d/%y")
```

```python
df["month"] = df["effective_to_date"].dt.month
```

```python
policies_state_month = (df.groupby(["state", "month"])["policy"].count().reset_index().pivot(index="state", columns="month", values="policy").astype(int))

policies_state_month
```

```python
#6.
top_states = (df.groupby("state")["policy"].count().sort_values(ascending=False).head(3).index)

top_states
```

```python
top_states_monthly = (
    df[df["state"].isin(top_states)]
    .groupby(["state", "month"])["policy"]
    .count()
    .reset_index()
    .pivot(index="state", columns="month", values="policy")
    .fillna(0)
    .astype(int))

top_states_monthly
```

```python
#7.
df["response_binary"] = df["response"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
melted = df.melt(
    id_vars=["sales_channel"],
    value_vars=["response_binary"],
    var_name="metric",
    value_name="value")
```

```python
response_rate = (melted
    .groupby("sales_channel")["value"]
    .mean()
    .round(3)
    .reset_index(name="response_rate"))

response_rate
```

The bonus analysis highlights clear geographic and channel-based patterns in policy sales and customer engagement. California, Oregon, and Arizona emerge as the top states in terms of total policies sold, accounting for a substantial share of overall sales. Across these states, policy sales remain relatively consistent between months, indicating stable demand during the observed period.

From a marketing perspective, the Agent channel demonstrates the highest customer response rate, outperforming Branch, Call Center, and Web channels. This suggests that more personalized, agent-driven interactions may be more effective in driving customer engagement.

Overall, these findings suggest that focusing marketing efforts on high-performing states and prioritizing agent-based channels could improve both policy sales and campaign effectiveness.
